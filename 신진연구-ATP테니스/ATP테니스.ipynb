```python
import pandas as pd
```

```python
scores_data = pd.read_csv('match_scores_2017.csv')

scores_data = scores_data.drop('tourney_year_id',axis=1)
scores_data = scores_data.drop('tourney_order',axis=1)
scores_data = scores_data.drop('tourney_slug',axis=1)
scores_data = scores_data.drop('tourney_url_suffix',axis=1)
scores_data = scores_data.drop('tourney_round_name',axis=1)
scores_data = scores_data.drop('round_order',axis=1)
scores_data = scores_data.drop('match_order',axis=1)
scores_data = scores_data.drop('winner_seed',axis=1)
scores_data = scores_data.drop('loser_seed',axis=1)
scores_data = scores_data.drop('match_score_tiebreaks',axis=1)
scores_data = scores_data.drop('winner_sets_won',axis=1)
scores_data = scores_data.drop('loser_sets_won',axis=1)
scores_data = scores_data.drop('winner_games_won',axis=1)
scores_data = scores_data.drop('loser_games_won',axis=1)
scores_data = scores_data.drop('winner_tiebreaks_won',axis=1)
scores_data = scores_data.drop('loser_tiebreaks_won',axis=1)
scores_data = scores_data.drop('match_stats_url_suffix',axis=1)

scores_data.head(3)
```

```python
stats_data = pd.read_csv('match_stats_2017.csv')
```

```python
winner = []; loser = []

for i in stats_data['match_id'].values:
    string = i.split('-')
    winner.append(string[2])
    loser.append(string[3])
```

```python
clean_stats_winner = pd.DataFrame()
clean_stats_winner['player_id'] = winner
clean_stats_winner['Aces'] = stats_data['winner_aces'] / (stats_data['winner_service_games_played'] + stats_data['winner_return_games_played'])
clean_stats_winner['Double Faults'] = stats_data['winner_double_faults'] / (stats_data['winner_service_games_played'] + stats_data['winner_return_games_played'])
clean_stats_winner['1st Serve'] = stats_data['winner_first_serves_in'] / stats_data['winner_first_serves_total']
clean_stats_winner['1st Serve Points Won'] = stats_data['winner_first_serve_points_won'] / stats_data['winner_first_serve_points_total']
clean_stats_winner['2nd Serve Points Won'] = stats_data['winner_second_serve_points_won'] / stats_data['winner_second_serve_points_total']
clean_stats_winner['Service Points Won'] = stats_data['winner_service_points_won'] / stats_data['winner_service_points_total']
clean_stats_winner['Break Points Saved'] = stats_data['winner_break_points_saved'] / stats_data['winner_break_points_serve_total']
clean_stats_winner['1st Serve Return Points Won'] = stats_data['winner_first_serve_return_won'] / stats_data['winner_first_serve_return_total']
clean_stats_winner['2nd Serve Return Points Won'] = stats_data['winner_second_serve_return_won'] / stats_data['winner_second_serve_return_total']
clean_stats_winner['Break Points Coverted'] = stats_data['winner_break_points_converted'] / stats_data['winner_break_points_return_total']
clean_stats_winner['Return Points Won'] = stats_data['winner_return_points_won'] / stats_data['winner_return_points_total']
clean_stats_winner['Result'] = 1
clean_stats_winner = clean_stats_winner.fillna(0)

clean_stats_winner.head(3)
```

```python
clean_stats_loser = pd.DataFrame()
clean_stats_loser['player_id'] = loser
clean_stats_loser['Aces'] = stats_data['loser_aces'] / (stats_data['loser_service_games_played'] + stats_data['loser_return_games_played'])
clean_stats_loser['Double Faults'] = stats_data['loser_double_faults'] / (stats_data['loser_service_games_played'] + stats_data['loser_return_games_played'])
clean_stats_loser['1st Serve'] = stats_data['loser_first_serves_in'] / stats_data['loser_first_serves_total']
clean_stats_loser['1st Serve Points Won'] = stats_data['loser_first_serve_points_won'] / stats_data['loser_first_serve_points_total']
clean_stats_loser['2nd Serve Points Won'] = stats_data['loser_second_serve_points_won'] / stats_data['loser_second_serve_points_total']
clean_stats_loser['Service Points Won'] = stats_data['loser_service_points_won'] / stats_data['loser_service_points_total']
clean_stats_loser['Break Points Saved'] = stats_data['loser_break_points_saved'] / stats_data['loser_break_points_serve_total']
clean_stats_loser['1st Serve Return Points Won'] = stats_data['loser_first_serve_return_won'] / stats_data['loser_first_serve_return_total']
clean_stats_loser['2nd Serve Return Points Won'] = stats_data['loser_second_serve_return_won'] / stats_data['loser_second_serve_return_total']
clean_stats_loser['Break Points Coverted'] = stats_data['loser_break_points_converted'] / stats_data['loser_break_points_return_total']
clean_stats_loser['Return Points Won'] = stats_data['loser_return_points_won'] / stats_data['loser_return_points_total']
clean_stats_loser['Result'] = 0
clean_stats_loser = clean_stats_loser.fillna(0)
clean_stats_loser.head(3)
```

```python
players_data = pd.read_csv('player_overviews.csv')

players_data = players_data.drop('first_name',axis=1)
players_data = players_data.drop('last_name',axis=1)
players_data = players_data.drop('player_url',axis=1)
players_data = players_data.drop('residence',axis=1)
players_data = players_data.drop('birthdate',axis=1)
players_data = players_data.drop('birthplace',axis=1)
players_data['handedness'] = players_data['handedness'].fillna('Unknown')
players_data['backhand'] = players_data['backhand'].fillna('Unknown')

players_data.head(3)
```

```python
rankings_data = pd.read_csv('rankings_1973-2017.csv')

clean_ranking = pd.DataFrame()
clean_ranking['year'] = rankings_data['week_year']
clean_ranking['rank'] = rankings_data['rank_number']
clean_ranking['age'] = rankings_data['player_age']
clean_ranking['point_per_played'] = rankings_data['ranking_points'] / rankings_data['tourneys_played']
clean_ranking['player_slug'] = rankings_data['player_slug']
clean_ranking['player_id'] = rankings_data['player_id']

clean_ranking.head(3)
```

```python
ranking_data = clean_ranking.dropna(0)
ranking_data.head(3)
```

```python
players_info = pd.merge(players_data, ranking_data, on='player_id', how='outer').dropna(axis=0).drop_duplicates(['player_id'])
players_info.head(3)
```

```python
stats_player = pd.concat([clean_stats_loser,clean_stats_winner])
stats_player.head(3)
```

```python
end_data = pd.merge(stats_player, players_info ,on='player_id', how='left').dropna(axis=0)
end_data.head(3)
```

```python
end_data.to_csv('2017_players_data_ML.csv', index=False)
```
