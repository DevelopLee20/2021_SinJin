```python
import pandas as pd
import numpy as np
```

```python
Data1 = pd.read_excel('twosum.xlsx', sheet_name='match_scores_1991-2016_unindexe')
Data2 = pd.read_excel('twosum.xlsx', sheet_name='match_stats_1991-2016_unindexed')
```

```python
Title1 = list(Data1)
Title2 = list(Data2)
```

```python
List1 = list(Data1.values)
List2 = list(Data2.values)
```

```python
data_list = []
```

```python
for i in List1:
    for j in List2:
        if i[0] == j[0]:
            ii = list(i)
            jj = list(j)
            data_list.append(ii+jj)
            break
```

```python
len(data_list)
```

```python
a = pd.DataFrame(data_list, columns=[Title1+Title2])
```
