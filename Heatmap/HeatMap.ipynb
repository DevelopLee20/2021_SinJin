```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/dldls/Desktop/vscoding/JoNel/Project_JoNel/Heatmap/Players_twosum.csv')
df.head(3)
```

```python
df = df.drop(['num'], axis=1)
df = df.drop(['winner_player_id'], axis=1)
df = df.drop(['loser_player_id'], axis=1)
df = df.drop(['match_id'], axis=1)
df = df.drop(['match_id.1'], axis=1)
df.head(3)
```

```python
col_lst = list(df.columns)
df.shape
```

```python
dff = df.values
dff = pd.DataFrame(dff, columns=col_lst)

cor = dff.corr()

ax = sns.heatmap(cor, annot=True)
```
