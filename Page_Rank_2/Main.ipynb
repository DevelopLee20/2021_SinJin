```python
import numpy as np
import pandas as pd
```

```python
excel_file = pd.read_excel('DreamTree.xlsx', sheet_name="2020_남고")
df_excel = pd.DataFrame(excel_file)
```

```python
df_excel.shape
```

```python
df_new = pd.DataFrame()
new_list = []

for i in df_excel['WINNER']:
    if i not in new_list:
        new_list.append(i)

df_new['Player'] = new_list

for i,j in zip(df_excel['WINNER'], df_excel['LOSER']):
    if j not in df_new:
        df_new[j] = 0

    idx = df_new[df_new['Player'] == i].index.tolist()
    df_new[j][idx[0]] = 1
```

```python
df_new['Count'] = df_new.sum(axis=1)
df_new
```
