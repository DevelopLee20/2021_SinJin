```python
import numpy as np
import pandas as pd
```

```python
excel_file = pd.read_excel('DreamTree.xlsx', sheet_name="2020_남고")
df_excel = pd.DataFrame(excel_file)
```

```python
df_excel.shape
```

```python
df_new = pd.DataFrame()
player_list = []

for i,j in zip(df_excel['WINNER'], df_excel['LOSER']):
    if i not in player_list:
        player_list.append(i)
    if j not in player_list:
        player_list.append(j)
```

```python
df_new['Player'] = player_list

for i in player_list:
    df_new[i] = 0
```

```python
df_new['Win'] = 0.0
df_new['Loss'] = 0.0
```

```python
for i,j in zip(df_excel['WINNER'], df_excel['LOSER']):
    idx = df_new[df_new['Player'] == i].index.tolist()
    df_new[j][idx] = 1
```

```python
df_new['Win'] = df_new.sum(axis=1)
```

```python
for i,j in zip(df_excel['WINNER'], df_excel['LOSER']):
    idx = df_new[df_new['Player'] == j].index.tolist()
    df_new['Loss'][idx] += 1
```

```python
df_new
```

```python
data = np.array(df_new)
```

```python
Team = df_new.shape[0]
damping_factor = 0.85
N = df_new.shape[0] - 1
pre_PageRank = 1.000000 / N
```

```python
result_list = []

for i in range(0,Team):
    result = 0

    for j in range(1,Team+1):
        if i+1 == j:
            pass
        if data[i][j] == 1:
            PR = pre_PageRank     # 페이지 랭크
            C = data[j-1][Team+2] # 총 링크의 개수
            result += PR / C

    result_list.append(((1-damping_factor)/N + damping_factor * result))

result_list
```

```python
dic = {}

for i in range(0,Team):
    dic[data[i][0]] = round(result_list[i],6)

dic = sorted(dic.items(), key= lambda x : x[1], reverse=True)
dic
```

```python
sum(result_list)
```
